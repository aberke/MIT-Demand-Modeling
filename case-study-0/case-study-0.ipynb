```python
""" 1.202 - Demand Modeling
Case Study 0
Alexandra Berke


VARIABLES:
"""

DUR ='DUR'
# Duration Dummy: 
#     1 if for long trips (over 3 weeks)
#     0 otherwise
# 80% of the population takes trips that are less than 3 weeks in duration.

CITY ='CITY'
# City Dummy
#     1 if travel to city
#     0 otherwise
# 30% of the population travel to cities for a vacation, 20% go camping and the rest pursue other recreational destinations.

CAMP = 'CAMP'
# Camping Dummy:  
#     1 if camping
#     0 otherwise

GRSIZE = 'GRSIZE'
# Size of travel group: 
# The size of a travel group is distributed with equal probability for each possible value between 1 and 9.


INC = 'INC'
# Income: Average disposable income of household to which traveler belongs (thousands of dollars)
# Disposable income for people who take summer vacations follows a (truncated) normal distribution with mean
# $25,000/year and a standard deviation of $7,000/year and is non-negative.
# mean: 25,000
# std:  7,000


EDUC = 'EDUC'
# Education: Education, measured by number of years after high school
# Education after high school is exponentially distributed with a mean of 5 years if income > $25000/year;
# otherwise it is distributed with a mean of 2 years.
# Data from an exponential distribution can be generated by a simple transformation of a uniform
# random variable. If R is a uniformly distributed random number between 0 and 1, and X is an exponentially
# distributed random variable with parameter l, then observations on X can be generated as - ln(R)/ l.


EXP = 'EXP'
# the following model is used for the calculation of vacation expenditure:
# EXP = 20.0 - 5.0DUR + 15.0CITY - l0.0CAMP - 2.5GRSIZE + 1.5INC + 2.5EDUC+ e (1)
# where e is normally distributed with a mean of 0 and a standard deviation of 5.0.

import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""Row generation."""


def generate_row():
    """Generates travel variables for each traveler and returns dictionary (row) representing dummy traveler:
    {DUR: 0/1, CITY: 0/1, CAMP: 0/1, GRSIZE: 0/1, INC: float, EDUC: int}
    
    Returns generated row.
    """
    row = dict()
    
    # Duration Dummy: DUR 
    # 1 if for long trips (over 3 weeks)
    # 0 otherwise
    # 80% of the population takes trips that are less than 3 weeks in duration.
    row[DUR] = 1 if (random.random() > 0.8) else 0
    
    # generate CITY vs CAMP vs neither
    # City Dummy:
    # 1 if travel to city
    # 0 otherwise
    # Camping Dummy:  
    # 1 if camping
    # 0 otherwise
    # 30% of the population travel to cities for a vacation,
    # 20% go camping and the rest pursue other recreational destinations.
    place = random.random()
    row[CITY] = 1 if (place <= 0.3) else 0
    row[CAMP] = 1 if (place > 0.8) else 0
    
    # Size of travel group: 
    # The size of a travel group is distributed with equal probability for each possible value between 1 and 9.
    row[GRSIZE] = random.choice(range(1, 10))
    
    # Disposable income for people who take summer vacations follows a (truncated) normal distribution with mean
    # $25,000/year and a standard deviation of $7,000/year and is non-negative.
    income = -1
    while income < 0:
        income = np.random.normal(25, 7, 1)[0]
    row[INC] = income

    # Education: Education, measured by number of years after high school
    # Education after high school is exponentially distributed with a mean of 5 years if income > $25000/year;
    # otherwise it is distributed with a mean of 2 years.
    # Data from an exponential distribution can be generated by a simple transformation of a uniform
    # random variable. If R is a uniformly distributed random number between 0 and 1, and X is an exponentially
    # distributed random variable with parameter l, then observations on X can be generated as - ln(R)/ l.
    education_mean = 5 if (income > 25) else 2
    l = 1/education_mean
    education = (-1*(np.log(random.random())/l))
    row[EDUC] = education
    
    return row
```

```python
"""Tasks"""
# 1. Generate the data for 200 observations of the independent variables using the procedure described in
# section 1.5 below.
observations = [generate_row() for i in range(200)]
observations_df = pd.DataFrame(observations)


# 2. Use the generated explanatory variables to generate data on expenditure using the model described in
# equation (1): EXP = 20.0 - 5.0DUR + 15.0CITY - l0.0CAMP - 2.5GRSIZE + 1.5INC + 2.5EDUC+ e
# where e is normally distributed with a mean of 0 and a standard deviation of 5.0.
observations_df[EXP] = 20.0 - 5.0*observations_df[DUR] + 15.0*observations_df[CITY] - observations_df[CAMP]*10.0 - 2.5*observations_df[GRSIZE] + 1.5*observations_df[INC] + 2.5*observations_df[EDUC]+ np.random.normal(0, 5, 200)

print('Task 3: Data for 100 observations')
# 3. Select the first 100 observations.
observations_100_df = observations_df.head(100)

observations_100_df
# a) Calculate the means, standard deviations, maxima and minima for EXP, DUR, INC, and EDUC.
# Also calculate the correlation or covariance matrix for these four variables.
task_3a = """3. a) Calculate the means, standard deviations, maxima and minima for EXP, DUR, INC, and EDUC.
Also calculate the correlation or covariance matrix for these four variables.
"""
print(task_3a)

for v in [EXP, DUR, INC, EDUC]:
    print(v)
    print('mean:', observations_100_df[v].mean())
    print('standard deviation:', observations_100_df[v].std())
    print('maxima:', observations_100_df[v].max())
    print('minima:', observations_100_df[v].min())
    print('\n')

print('Correlation matrix:')
print(observations_100_df[[EXP, DUR, INC, EDUC]].corr())
```

```python
task_3c = """3. c) Given this sample, test the hypothesis that the mean income of the (sub) population from which
this sample was drawn is significantly different from $25,000 (the alleged mean income of the
entire population). Explain the reasoning behind the test-statistic that you use. Please do not use
excel functions but show your own calculations."""
print(task_3c)

task_3c_explanation = """
I use a type 1 error test.  I use a two-tailed test because I do not care if my data
values are above or below the poluation and I want to test for significant difference in either direction.
I use a Type 1 error test in order to test the null hypothesis that the mean is $25,000 because
I have prior knowledge and therefore a strong a priori belief that the sub population was
drawn from a distribution with mean income $25,000.

I define: delta = (alleged population mean for income) - (observed mean for income)

I use null hypothesis H0: delta = 0
alternative hypothesis H1: delta != 0

Test statistic: (delta_hat/s_hat), where s_hat is the observed standard deviation.
Define (delta_critical/s_hat), where delta_critical determines the value such that the null hypothesis should be rejected.

To determined delta_critical, I use significance level alpha = 0.05.
==> delta_critical/s_hat = +/- 1.65
==> delta_critical = (s_hat)*(1.65)

"""
print(task_3c_explanation)

observed_mean = observations_100_df[INC].mean()
print('observed_mean:', observed_mean)

s_hat = observations_100_df[INC].std()
print('s_hat:', s_hat)

delta_hat = 25.0 - observed_mean
print('delta_hat:', delta_hat)


delta_critical = s_hat*(1.65)
print('delta_critical: +/-', delta_critical)

task_3c_explanation_result = """
Because delta_hat is well within the range of confidence (i.e. closer to 0
than delta_critical), I will not reject the null hypothesis: The mean income of the sub population is not
significantly different than the alleged income mean of $25,000.
"""
print(task_3c_explanation_result)
```
